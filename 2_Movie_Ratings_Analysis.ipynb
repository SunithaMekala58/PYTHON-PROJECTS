# Outlier Detection 

```python
# Get the first quantile
Q1 = movies['AudienceRating'].quantile(0.25)
# Get the third quantile
Q3 = movies['AudienceRating'].quantile(0.75)

# IQR - Inter Quantile Range
IQR = Q3 - Q1

#define bounds
lower = Q1-1.5*IQR
upper = Q3+1.5*IQR

outlier = movies[(movies["AudienceRating"]<lower) | 
                 (movies["AudienceRating"]>upper)]
print("Outliers:\n", outlier)
```

## How can you style your dashboard  using different color map

```python
# python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)

sns.set_style('dark',{'axes.facecolor':'black'})
f, axes = plt.subplots (2,2, figsize = (15,15))

#plot [0,0]
k1 = sns.kdeplot(x= movies.BudgetMillions,y = movies.AudienceRating, \
                 shade = True, shade_lowest=True,cmap = 'BuPu', \
                 ax = axes[0,0])
k1b = sns.kdeplot(x = movies.BudgetMillions,y = movies.AudienceRating, \
                 cmap = 'viridis',ax = axes[0,0])

#plot [0,1]
k2 = sns.kdeplot(x = movies.BudgetMillions,y = movies.CriticRating,\
                 shade=True, shade_lowest=True, cmap='cividis',\
                 ax = axes[0,1])
k2b = sns.kdeplot(x = movies.BudgetMillions,y = movies.CriticRating,\
                  cmap = 'magma', ax = axes[0,1])

#plot[1,0]
z = sns.violinplot(data=movies[movies.Genre=='Drama'], \
                   x='Year', y = 'CriticRating', ax=axes[1,0], palette = 'Set1')

#plot[1,1]
k4 = sns.kdeplot(x = movies.CriticRating,y = movies.AudienceRating, \
                 shade = True,shade_lowest=False,cmap='PuBuGn', \
                 ax=axes[1,1])

k4b = sns.kdeplot(x = movies.CriticRating, y = movies.AudienceRating, \
                  cmap='YlOrBr',ax = axes[1,1])


k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))

plt.show()
```
